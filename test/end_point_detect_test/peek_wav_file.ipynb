```python
peek_wav_idx = 13;
peek_wav_file = single_fold_alarm_list[peek_wav_idx];
print(f"peek wav name:{peek_wav_file}");
```

```python
wavobj = wave.open(peek_wav_file, 'rb') #use wave to
```

```python
print('Parameters:', wavobj.getparams())
```

```python
sample_freq = wavobj.getframerate();#Sample frequency, this is the number of samples per second
n_samples = wavobj.getnframes();#Number of samples, this is the total number of samples or frames in the audio file
signal_wave = wavobj.readframes(-1);#Signal wave, this is the wave amplitude which is the sound intensity.
duration = n_samples/sample_freq;#Audio length, this is the duration of the audio.
print(f"sample_freq:{sample_freq}");
print(f"n_samples:{n_samples}");
# print(f"singal_wave:{signal_wave}");
print(f"duration:{duration}");
```

```python
wav_array = np.frombuffer(signal_wave, dtype=np.int16)
time = np.linspace(0, duration, num=n_samples)
```

```python
def plot_wav(wavsig):
    plt.plot(wavsig)
    plt.ylabel("Amplitude")
    plt.xlabel("Time")
    plt.show()
```

```python
IPython.display.display(Audio(wav_array,rate=20000))
plot_wav(wav_array)
```
